```python
# Logistic Regression Concepts Notebook

# Question 1: Linear vs Logistic Regression
"""
Key Differences:
1. Output Type:
   - Linear: Continuous values (e.g., house prices)
   - Logistic: Probabilities (0-1) for classification

2. Equation:
   - Linear: y = β₀ + β₁x₁ + ... + βₙxₙ
   - Logistic: p = 1/(1 + e^-(β₀ + β₁x₁ + ... + βₙxₙ))

3. Best Fit:
   - Linear: Least squares
   - Logistic: Maximum likelihood

Example for Logistic:
Predicting whether a patient has diabetes (Yes/No) based on glucose levels, BMI, and age.
"""

# Question 2: Logistic Regression Cost Function
"""
Cost Function: Binary Cross-Entropy
J(θ) = -1/m ∑ [yⁱ log(hθ(xⁱ)) + (1-yⁱ) log(1-hθ(xⁱ))]

Optimization:
1. Typically minimized using gradient descent
2. Updates parameters θ iteratively:
   θⱼ := θⱼ - α ∂J(θ)/∂θⱼ
3. α = learning rate
4. Convex function guarantees global minimum
"""

# Question 3: Regularization in Logistic Regression
"""
Regularization Types:
1. L1 (Lasso): Adds λ∑|θⱼ| to cost function
2. L2 (Ridge): Adds λ∑θⱼ² to cost function

Prevents Overfitting by:
1. Penalizing large coefficients
2. Reducing model complexity
3. Handling multicollinearity
4. Performing feature selection (L1)
"""

# Question 4: ROC Curve Evaluation
"""
ROC Curve:
- Plots True Positive Rate (TPR) vs False Positive Rate (FPR)
- At various classification thresholds
- AUC (Area Under Curve) measures overall performance

Interpretation:
- AUC = 1: Perfect classifier
- AUC = 0.5: Random guessing
- AUC > 0.8: Good model

Usage:
1. Compare models
2. Choose optimal threshold
3. Evaluate trade-offs between sensitivity/specificity
"""
```
