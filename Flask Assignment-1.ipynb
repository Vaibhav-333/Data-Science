```python
# Flask Framework Notebook - Concise Answers

# Question 1: What is Flask Framework? Advantages
"""
Flask: Lightweight Python web framework for building web applications.

Advantages:
1. Simple & lightweight
2. Flexible structure
3. Built-in development server
4. Easy to learn
5. Extensible with plugins
"""

# Question 2: Hello World Flask App
from flask import Flask
app = Flask(__name__)

@app.route('/')
def hello():
    return 'Hello World!!'

if __name__ == '__main__':
    app.run(debug=True)
    
# [Attach screenshot of "Hello World!!" output in browser here]
# Output: Displays "Hello World!!" at http://127.0.0.1:5000/

# Question 3: App Routing in Flask
"""
App routing: Mapping URLs to Python functions using @app.route().

Why use routes:
1. Direct URLs to specific functions
2. Create clean URL structure
3. Handle different HTTP methods
4. Organize application logic
"""

# Question 4: Welcome and Company Routes
from flask import Flask
app = Flask(__name__)

@app.route('/welcome')
def welcome():
    return 'Welcome to ABC Corporation'

@app.route('/')
def company_details():
    return '''
    Company Name: ABC Corporation<br>
    Location: India<br>
    Contact Detail: 999-999-9999
    '''

if __name__ == '__main__':
    app.run(debug=True)
    
# Outputs:
# /welcome -> "Welcome to ABC Corporation"
# / -> Shows company details in 3 lines
```
