```python
# Forward Propagation in Neural Networks Notebook

# Question 1: Purpose of forward propagation
"""
Forward propagation is the process of passing input data through the neural network to generate predictions. It:
1. Computes the network's output
2. Transforms inputs through successive layers
3. Enables comparison with true values (for loss calculation)
4. Is essential for both training and inference
"""

# Question 2: Mathematical implementation in single-layer network
"""
For a single-layer network with input x, weights W, bias b, and activation f:
z = W·x + b  # Weighted sum
a = f(z)     # Activation output
Where:
- x ∈ ℝⁿ (input vector)
- W ∈ ℝᵐˣⁿ (weight matrix)
- b ∈ ℝᵐ (bias vector)
- f: activation function
"""

# Question 3: Activation functions in forward propagation
"""
Activation functions:
1. Introduce non-linearity between layers
2. Transform weighted sums (z) to layer outputs (a)
3. Enable learning complex patterns
4. Applied element-wise to z values
Example: a = ReLU(W·x + b)
"""

# Question 4: Role of weights and biases
"""
Weights (W):
1. Determine connection strength between neurons
2. Scalar multipliers for input features
3. Learned during training

Biases (b):
1. Allow shifting activation function output
2. Enable fitting when all inputs are zero
3. Provide additional learnable parameters
"""

# Question 5: Softmax in output layer
"""
Softmax function:
1. Converts output scores to probabilities
2. Ensures outputs sum to 1 (for classification)
3. Highlights largest values while suppressing others
Formula for class j: σ(z)ⱼ = eᶻʲ / ∑ₖ eᶻᵏ
Used for:
- Multi-class classification
- Interpretable probability outputs
"""
```
